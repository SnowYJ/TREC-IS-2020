### use BeautifulSoup to extract title behind link

```python
import pandas as pd

raw_dataset = pd.read_csv("pre-processed data/raw_label_dataset.csv", lineterminator="\n")
```

```python
raw_dataset_critical = raw_dataset[raw_dataset['priority'] == 'Critical']

raw_dataset_critical = raw_dataset_critical.reset_index(drop=True) 
```

```python
raw_dataset_critical[:3]
```

```python
import re

url_pattern =  re.compile(r'''(?i)\b((?:https?://|www\d{0,3}[.]|[a-z0-9.\-]+[.][a-z]{2,4}/)(?:[^\s()<>]+|\(([^\s()<>]+|(\([^\s()<>]+\)))*\))+(?:\(([^\s()<>]+|(\([^\s()<>]+\)))*\)|[^\s`!()\[\]{};:'".,<>?«»“”‘’]))''')
```

```python
link_dict = {}

for i in range(len(raw_dataset_critical)):
    content = raw_dataset_critical.loc[i, 'content']
    result = re.findall(url_pattern, str(content))
    if len(result) == 1: 
        link_dict[i] = result[0][0]
    if len(result) == 2: 
        link_dict[i] = [result[0][0], result[1][0]]
        
```

```python
print("How many link in critical tweets: ", len(link_dict))
```

```python
import requests
import bs4
from bs4 import BeautifulSoup
import time
from tqdm.notebook import tqdm
import ipaddress
import socket
import eventlet
IP4=str(ipaddress.ip_address(8888))
IPAddr = socket.gethostbyname(IP4)
```

```python
def get_title(url):
    status = 1
    connect_times = 0
    eventlet.monkey_patch()
    
    # set the maximum time.
    with eventlet.Timeout(30, False):
        while status == 1 and connect_times <= 3:
            try:
                response = requests.get(url, headers={"Connection": "close"})
                status = 0
            except:
                status = 1
                connect_times += 1
                time.sleep(5)

        if status == 1:
            return (0, 'None')
        else:
            status_code = response.status_code
            if status_code == 200:
                content = bs4.BeautifulSoup(response.content.decode("utf-8"), 'html.parser')
                if content.title != None:
                    res =  (status_code, content.title.string)
                else:
                    res =  (status_code, 'None')
            else:
                res = (status_code, 'None')
            return res
        
    return (0, 'None')
```
